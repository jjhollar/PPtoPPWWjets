```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Old way of cutting, with bands on both sides
band1_cut = np.array([0.5,1,1.5,2])
band2_cut = np.array([0.5,1,1.5,2])

sig_A = np.array([21,78,109,125])
sig_B = np.array([146,179,185,190])

antiPruned_A = np.array([2,8,16,28])
antiPruned_B = np.array([62,121,171,213])

sig = np.add(sig_A,sig_B)
antiPruned = np.add(antiPruned_A,antiPruned_B)

baseline = np.ones_like(sig) * 120/np.sqrt(22)
```

```python
plt.figure(figsize=(15,10))
plt.plot(band1_cut, sig/np.sqrt(antiPruned),label="Merging area A and B")
plt.plot(band1_cut, sig_A/np.sqrt(antiPruned_A),label="Signal region A")
plt.plot(band1_cut, sig_B/np.sqrt(antiPruned_B),label="Signal region B")
plt.plot(band1_cut, baseline,label="Previous cut")
plt.xlabel(r"n$_\sigma$ band widths",fontsize=20)
plt.ylabel(r"$\frac{S}{\sqrt{B}}$",fontsize=20,rotation=0,labelpad=30)
plt.legend()
plt.show()
```
